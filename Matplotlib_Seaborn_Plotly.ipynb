```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
titanic_df=pd.read_csv("titanic-training-data.csv")
```

```python

import matplotlib.colors as mcolors
# Get all named colors
colors = mcolors.CSS4_COLORS
colors
# Print all color names
for name, hex in colors.items():
    print(name, hex)
```

```python
### Matplotlib
```

```python
# Count the number of passengers in each passenger class
class_counts = titanic_df['Pclass'].value_counts()

# Set custom colors for the bars
colors = ['#2b8cbe', '#a6bddb', "#778899"]

# Set the figure size
plt.figure(figsize=(8, 6))

# Create bar chart of passenger counts by passenger class
plt.bar(class_counts.index, class_counts.values, color=colors)

# Set the title and axis labels
plt.title('Passenger Counts by Class')
plt.xlabel('Class')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Count the number of passengers who survived and did not survive
survival_counts = titanic_df['Survived'].value_counts()

# Set custom colors for the slices
colors = ['#2b8cbe', '#a6bddb']

# Set the figure size
plt.figure(figsize=(8, 6))

# Create pie chart of passenger survival
plt.pie(survival_counts.values, labels=survival_counts.index, colors=colors, autopct='%1.1f%%')

# Set the title
plt.title('Passenger Survival')

# Show the plot
plt.show()
```

```python
# Set custom colors for the points
colors = {0: '#a6bddb', 1: '#2b8cbe'}

# Set the figure size
plt.figure(figsize=(8, 6))

# Create scatter plot of passenger age vs. fare, colored by survival status
plt.scatter(titanic_df['Age'], titanic_df['Fare'], c=titanic_df['Survived'].map(colors))

# Set the title and axis labels
plt.title('Passenger Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')

# Show the plot
plt.show()
```

```python
# Group data by Class, Gender, and Survival
grouped_data = titanic_df.groupby(['Pclass', 'Sex', 'Survived']).size().unstack()

# Set custom colors for the bars
colors = ['#a6bddb', '#2b8cbe']

# Set the figure size
plt.figure(figsize=(8, 6))

# Create stacked bar chart of survival rates by Class and Gender
grouped_data.plot(kind='bar', stacked=True, color=colors)

# Set the title and axis labels
plt.title('Survival Rates by Class and Gender')
plt.xlabel('Class and Gender')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Count the number of passengers in each passenger class
class_counts = titanic_df['Pclass'].value_counts()

# Set custom colors for the slices
colors = ['#2b8cbe', '#a6bddb', '#ece7f2']

# Set the figure size
plt.figure(figsize=(8, 6))

# Create pie chart of passenger class proportions
plt.pie(class_counts.values, labels=class_counts.index, colors=colors, autopct='%1.1f%%', startangle=90)

# Set the title
plt.title('Passenger Class Proportions')

# Show the plot
plt.show()
```

```python
import numpy as np
```

```python
# Group data by Age and Fare and compute the survival rate
grouped_data = titanic_df.groupby([pd.cut(titanic_df['Age'], np.arange(0, 100, 10)), pd.cut(titanic_df['Fare'], np.arange(0, 600, 50))])['Survived'].mean().unstack()

# Set the figure size
plt.figure(figsize=(8, 6))

# Create heatmap of survival by age and fare
plt.imshow(grouped_data.values, cmap='RdYlGn', interpolation='nearest', aspect='auto', vmin=0, vmax=1)

# Set the x-axis and y-axis ticks and labels
plt.xticks(np.arange(len(grouped_data.columns)), grouped_data.columns.categories, rotation=45)
plt.yticks(np.arange(len(grouped_data.index)), grouped_data.index.categories)

# Add colorbar
plt.colorbar()

# Set the title and axis labels
plt.title('Survival Rate by Age and Fare')
plt.xlabel('Fare')
plt.ylabel('Age')

# Show the plot
plt.show()
```

```python
# Group by gender and survived
gender_survived = titanic_df.groupby(['Sex', 'Survived']).size().unstack()

# Create horizontal bar chart
gender_survived.plot(kind='barh', stacked=True, color=['#2b8cbe', '#a6bddb'], figsize=(8, 6))

# Set the title and axis labels
plt.title('Survived Passengers by Gender')
plt.xlabel('Number of Passengers')
plt.ylabel('Gender')

# Show the plot
plt.show()
```

```python
# Group by age and class
age_class = titanic_df.groupby(['Age', 'Pclass']).size().unstack()

# Create stacked area chart
age_class.plot(kind='area', stacked=True, color=['#2b8cbe', '#a6bddb', '#ece7f2'], figsize=(8, 6))

# Set the title and axis labels
plt.title('Passenger Age Distribution by Class')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Create countplot of passengers by class
sns.countplot(x='Pclass', data=titanic_df)

# Set the title and axis labels
plt.title('Passengers by Class')
plt.xlabel('Class')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Create violinplot of age by class
sns.violinplot(x='Pclass', y='Age', data=titanic_df)

# Set the title and axis labels
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')

# Show the plot
plt.show()
```

```python

# Compute the correlation matrix
corr_matrix = titanic_df.corr()

# Set the figure size
plt.figure(figsize=(8, 6))

# Create heatmap of correlation matrix
sns.heatmap(corr_matrix, annot=True)

# Set the title and axis labels
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(5,5))

# Create swarmplot of fares by class
sns.swarmplot(x='Pclass', y='Fare', data=titanic_df)

# Set the title and axis labels
plt.title('Fares by Class')
plt.xlabel('Class')
plt.ylabel('Fare')

# Show the plot
plt.show()
```

```python
sns.set_style('whitegrid')
sns.displot(data=titanic_df, x='Fare', hue='Pclass', palette='pastel', kde=True)
plt.title('Distribution of Fare by Class')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python


# Create heatmap of passenger survival rate by passenger class and age
fig = px.density_heatmap(titanic_df, x='Age', y='Pclass', z='Survived', marginal_x='histogram')

# Show the plot
fig.show()
```

```python
# Create treemap of passenger class and survival rate
class_survival_counts = titanic_df.groupby(['Pclass', 'Survived']).size().reset_index(name='Count')
fig = px.treemap(class_survival_counts, path=['Pclass', 'Survived'], values='Count')

# Show the plot
fig.show()



```

```python
import plotly.express as px
from sklearn.datasets import load_iris

# Load iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= iris['feature_names'] + ['target'])

# Create 3D scatter plot
fig = px.scatter_3d(iris_df, x='sepal length (cm)', y='sepal width (cm)', z='petal length (cm)',
              color='target', symbol='target')

# Show the plot
fig.show()
```

```python
# Create 3D scatter plot
fig = px.scatter_3d(titanic_df, x='Age', y='Fare', z='Survived',
              color='Pclass', symbol='Sex')

# Show the plot
fig.show()
```

```python
# Create scatter matrix of passenger age, fare, and number of siblings/spouses
fig = px.scatter_matrix(titanic_df, dimensions=['Age', 'Fare', 'SibSp'], color='Survived')

# Show the plot
fig.show()
```

```python
import plotly.express as px
import pandas as pd


# Create bubble chart of passenger age, fare, and survival rate
fig = px.scatter(titanic_df, x='Age', y='Fare', size='Survived', color='Survived', hover_data=['Name'])

# Show the plot
fig.show()
```

```python
# Create sunburst chart of passenger survival by passenger class, gender, and age
class_gender_age_counts = titanic_df.groupby(['Pclass', 'Sex', pd.cut(titanic_df.Age, bins=[0, 18, 30, 60, 100])]).Survived.sum().reset_index()
class_gender_age_counts['Age'] = class_gender_age_counts.Age.astype(str)
fig = px.sunburst(class_gender_age_counts, path=['Pclass', 'Sex', 'Age'], values='Survived')

# Show the plot
fig.show()
```

```python
# Create heatmap of passenger survival by passenger class and sex
class_sex_counts = titanic_df.groupby(['Pclass', 'Sex']).Survived.sum().reset_index()
fig = px.imshow(class_sex_counts.pivot(index='Sex', columns='Pclass', values='Survived'))

# Show the plot
fig.show()
```
